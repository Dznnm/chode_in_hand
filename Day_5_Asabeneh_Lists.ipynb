```python
#len() for length of a list
fruits=['apple','banana','cherry','lemon']
print('length of fruits:',len(fruits))
vegetables=['tomato','potato','cabbage','onion','carrot']
animal_products=['milk','meat','butter','yoghurt']
web_techs=['HTML','CSS','JS','React','Redux','Node','MongoDB']
countries=['Finland','Estonia','Sweden','Denmark','Norway']

#Print lists and length
print('List of fruits:',fruits)
print('Number of fruits:',len(fruits))
print('List of vegetables:',vegetables)
print('Number of vegetables:',len(vegetables))
print('List of anime products:',animal_products)
print('Number of animal products:',len(animal_products))
print('List of web technologies:',web_techs)
print('Number of web technologies:',len(web_techs))
print('List of Countries:',countries)
print('Number of countries:',len(countries))
```

```python
#Accessing list items using positive indexing
print(fruits)
first_fruit=fruits[0]
print(first_fruit)
second_fruit=fruits[1]
last_fruit=fruits[len(fruits)-1]
print(second_fruit,last_fruit)

#Negative Indexing -1 is last, -2 second last
print(fruits)
last_fruit=fruits[-1]
second_last_fruit=fruits[-2]
print(last_fruit,second_last_fruit)
first_fruit=fruits[-len(fruits)]
print(first_fruit)

#unpacking list items
items=['item1','item2','item3','item4','item5']
first_item,second_item,third_item,*rest=items
print(first_item,second_item,third_item)
print(rest)
#example
countries=['Germany','France','Belgium','Sweden','Denmark','Finland','Norway','Iceland','Estonia']
gr,fr,bg,*scandic,es=countries
print(gr,fr,bg)
print(scandic)
```

```python
#Slicing items from a list positive indexing
fruits=['banana','orange','mango','lemon']
print(fruits)
all_fruits=fruits[0:4]
print(all_fruits)
orange_mango=fruits[1:3]
print(orange_mango)
orange_mango_lemon=fruits[1:]
print(orange_mango_lemon)
orange_and_lemon=fruits[::2]    #will take every 2nd item
print(orange_and_lemon)
```

```python
#negative
all_fruits=fruits[-4:]
orange_and_mango=fruits[-3:-1]
orang_mango_lemon=fruits[-3:]
reverse_fruits=fruits[::-1]
print(all_fruits)
print(orange_and_mango)
print(orang_mango_lemon)
print(reverse_fruits)
```
