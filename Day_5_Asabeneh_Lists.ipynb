```python
#List
empty_list=list()
print('length of empty list: ',len(empty_list))
empty_2=[]
print('length of empty 2:',len(empty_2))
```

```python
#len() for length of a list
fruits=['apple','banana','cherry','lemon']
print('length of fruits:',len(fruits))
vegetables=['tomato','potato','cabbage','onion','carrot']
animal_products=['milk','meat','butter','yoghurt']
web_techs=['HTML','CSS','JS','React','Redux','Node','MongoDB']
countries=['Finland','Estonia','Sweden','Denmark','Norway']

#Print lists and length
print('List of fruits:',fruits)
print('Number of fruits:',len(fruits))
print('List of vegetables:',vegetables)
print('Number of vegetables:',len(vegetables))
print('List of anime products:',animal_products)
print('Number of animal products:',len(animal_products))
print('List of web technologies:',web_techs)
print('Number of web technologies:',len(web_techs))
print('List of Countries:',countries)
print('Number of countries:',len(countries))
```

```python
#lists can differ in types
about=['Dzannun',300,True,{'country':'Indonesia','city':'Jombang'}]
print(about)
```

```python
#Accessing list items using positive indexing
print(fruits)
first_fruit=fruits[0]
print(first_fruit)
second_fruit=fruits[1]
last_fruit=fruits[len(fruits)-1]
print(second_fruit,last_fruit)

#Negative Indexing -1 is last, -2 second last
print(fruits)
last_fruit=fruits[-1]
second_last_fruit=fruits[-2]
print(last_fruit,second_last_fruit)
first_fruit=fruits[-len(fruits)]
print(first_fruit)

#unpacking list items
items=['item1','item2','item3','item4','item5']
first_item,second_item,third_item,*rest=items
print(first_item,second_item,third_item)
print(rest)
#example
countries=['Germany','France','Belgium','Sweden','Denmark','Finland','Norway','Iceland','Estonia']
gr,fr,bg,*scandic,es=countries
print(gr,fr,bg)
print(scandic)
```

```python
#Slicing items from a list positive indexing
fruits=['banana','orange','mango','lemon']
print(fruits)
all_fruits=fruits[0:4]
print(all_fruits)
orange_mango=fruits[1:3]
print(orange_mango)
orange_mango_lemon=fruits[1:]
print(orange_mango_lemon)
orange_and_lemon=fruits[::2]    #will take every 2nd item
print(orange_and_lemon)
```

```python
#negative
all_fruits=fruits[-4:]
orange_and_mango=fruits[-3:-1]
orang_mango_lemon=fruits[-3:]
reverse_fruits=fruits[::-1]
print(all_fruits)
print(orange_and_mango)
print(orang_mango_lemon)
print(reverse_fruits)
```

```python
#modifying Lists
fruits=['banana','orange','mango','lemon']
fruits[0]='avocado'
print(fruits)
fruits[1]='apple'
print(fruits)
last_index=len(fruits)-1
fruits[last_index]='lime'
print(fruits)
```

```python
#Checking items in a list
fruits=['banana','orange','mango','lemon']
exists='banana'in fruits
print(exists)
exists='lime'in fruits
print(exists)
```

```python
#joining lists
#Plus operator
positive_num=[1,2,3,4,5]
zero=[0]
negative_num=[-5,-4,-3,-2,-1]
integers=negative_num+zero+positive_num
print(integers)
fruits=['banana','orange','mango','lemon']
vegetables=['tomato','potato','cabbage','onion','carrot']
fruits_and_vegetables=fruits+vegetables
print(fruits_and_vegetables)
```

```python
#Extend
num1=[0,1,2,3]
num2=[4,5,6]
num1.extend(num2)
print('Numbers:',num1)
print(negative_num)
print(positive_num)
negative_num.extend(zero)
negative_num.extend(positive_num)
print('Integers:',negative_num)
fruits.extend(vegetables)
print('Fruits and Vegetables:',fruits)
```

```python
#counting items in a list
fruits=['banana','orange','mango','lemon']
print(fruits.count('orange'))
ages=[22,19,24,25,26,24,25,24]
print(ages.count(24))
```

```python
#Finding index of an item
print(fruits)
print(fruits.index('orange'))
print(ages)
print(ages.index(24))
```

```python
#reverse order of list
print(fruits)
fruits.reverse()
print(fruits)
print(ages)
ages.reverse()
print(ages)
```

```python
#sorting list items
print(fruits)
fruits.sort()
print(fruits)
fruits.sort(reverse=True)
print(fruits)
print(ages)
ages.sort()
print(ages)
ages.sort(reverse=True)
print(ages)
```

```python
fruits=['banana','orange','mango','lemon']
print(fruits)
print(sorted(fruits))
print(sorted(fruits,reverse=True))
print(fruits)
```
