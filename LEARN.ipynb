```python
#Hitung Pajak
print('Gaji Bruto =',gaji_total,'rupiah')
BPJS=220000
bruto_tahun=gaji_total*12
biaya_jabatan_tahun=6000000
JHT=110000
JP=55000
kes=55000
BPJS_pengurang_tahun=JHT*12+JP*12
pengurang_tahun=BPJS_pengurang_tahun+biaya_jabatan_tahun
netto_tahun=bruto_tahun-pengurang_tahun-54000000
print('netto setahun =',netto_tahun)
pajak_60jt=0.05*60000000
print('pajak dari 60jt setahun =',pajak_60jt)
pajak_sisa60jt=0.15*(netto_tahun-60000000)
print('pajak setahun after60jt =',pajak_sisa60jt)
pajak_tahun=pajak_60jt+pajak_sisa60jt
print('pajak setahun =',pajak_tahun)
pajak_bulan=pajak_tahun/12
print('Pajak bulan ini =',pajak_bulan)
gaji_bersih=gaji_total-pajak_bulan-JHT-JP-kes
print('Gaji bersih bulan depan =',gaji_bersih)
```

```python
#Assessing Characters in Strings by Index (counting from = 0)
language='Python'
first_letter=language[0]
print(first_letter)
second_letter=language[1]
last_index=len(language)-1
last_letter=language[last_index]
print(second_letter)
print(last_letter)
```

```python
#Starting from right end (-1) is the last index
language='Python'
last_letter=language[-1]
print(last_letter)
second_last=language[-2]
print(second_last)
```

```python
#Slicing strings into sub-strings
language='Python'
first_three=language[0:3]
print(first_three)
last_three=language[3:6]
print(last_three)
#another way
last_three=language[-3:]
print(last_three)
last_three=language[3:]
print(last_three)
```
