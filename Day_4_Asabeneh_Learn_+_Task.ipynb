```python
#Perkiraan Gaji B depan
hari_kerja=21
hari_merah=7
hari_long=21
gaji_pokok=5500000
rupiah_per_jam=gaji_pokok/22/8
jam_long=21*4.5
jam_merah=7*25
jam_kerja=8*22
jam_total=jam_kerja+jam_merah+jam_long
print(jam_total)
gaji_jam=jam_total*rupiah_per_jam
print(gaji_jam)
gaji_transport=30000*21
gaji_kpi=0.1*gaji_pokok
gaji_hadir=200000
gaji_shift=15000*14
gaji_total=gaji_jam+gaji_transport+gaji_kpi+gaji_hadir+gaji_shift
print('gaji total =',gaji_total)
#
```

```python
#Hitung Pajak
print('Gaji Bruto =',gaji_total,'rupiah')
BPJS=220000
bruto_tahun=gaji_total*12
biaya_jabatan_tahun=6000000
JHT=110000
JP=55000
kes=55000
BPJS_pengurang_tahun=JHT*12+JP*12
pengurang_tahun=BPJS_pengurang_tahun+biaya_jabatan_tahun
netto_tahun=bruto_tahun-pengurang_tahun-54000000
print('netto setahun =',netto_tahun)
pajak_60jt=0.05*60000000
print('pajak dari 60jt setahun =',pajak_60jt)
pajak_sisa60jt=0.15*(netto_tahun-60000000)
print('pajak setahun after60jt =',pajak_sisa60jt)
pajak_tahun=pajak_60jt+pajak_sisa60jt
print('pajak setahun =',pajak_tahun)
pajak_bulan=pajak_tahun/12
print('Pajak bulan ini =',pajak_bulan)
gaji_bersih=gaji_total-pajak_bulan-JHT-JP-kes
print('Gaji bersih bulan depan =',gaji_bersih)
```
