```python
empty_tuple=tuple()
tpl=('item1','item2','item3')
fruits=('banana','orange','mango','lemon')
print('Example of tuples:',tpl,fruits)
print('Length of tuple fruits:',len(fruits))
```

```python
#positive indexing similar to list
print('fruits:',fruits)
first_fruit=fruits[0]
print('The first fruit:',first_fruit)
last_index=len(fruits)-1
last_fruit=fruits[last_index]
print('The last fruit:',last_fruit)
```

```python
#negative indexing similar to list
print('Fruits tuple:',fruits)
first_fruit=fruits[-4]
second_fruit=fruits[-3]
last_fruit=fruits[-1]
print('The first fruit:',first_fruit)
print('The second fruit:',second_fruit)
print('The last fruit:',last_fruit)
```

```python
#slicing tuples
print("Fruits tuple:",fruits)
all_fruits=fruits[0:]
print(all_fruits)
middle_two_fruits=fruits[1:3]
print(middle_two_fruits)
```

```python
all_fruits=fruits[-4:]
print(all_fruits)
middle_two=fruits[-3:-1]
print(middle_two)
```

```python
#tuple to list
print('fruits tuple:',fruits)
fruits=list(fruits)
print('fruits list:',fruits)
fruits[0]='apple'
fruits=tuple(fruits)
print(fruits)
```

```python
print(fruits)
print('orange' in fruits)
print('kiwi' in fruits)
print(fruits[0]=='banana')
```

```python
print(tpl,fruits)
all_tuple=tpl+fruits
print(all_tuple)
```
